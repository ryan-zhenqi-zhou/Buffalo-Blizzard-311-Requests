```python
requests = pd.read_csv("../data/311_data_19_15_filtered.csv")
```

```python
groupby_type = requests.groupby(["category"]).agg({"Case Reference":"count"}).reset_index()
groupby_type2 = groupby_type.sort_values(by = ["Case Reference"], ascending = False).reset_index(drop=True)
category_2 = ["Requests related to snow removal","Requests related to garbage/recycling collection","Requests related to roads and parking","Requests related to trees","Requests related to housing",
              "Other requests", "Requests related to animals and pests", "Requests related to water and sewer"]
groupby_type2['category_2'] = category_2

groupby_type2
```
