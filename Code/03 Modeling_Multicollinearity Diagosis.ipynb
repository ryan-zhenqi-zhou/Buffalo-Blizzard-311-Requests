```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import math
import os
from scipy import stats
```

```python
def standarize_data(data, stats):
    return (data - stats['mean'])/ stats['std']
```

```python
data = pd.read_csv("../02 data/modeling.csv") 
```

```python
data.head()
```

```python
del data['311_index_per_property']
data = data.set_index("Geography")
```

```python
X = data.copy()
```

```python
#standardization
X_stats = X.describe().transpose()
X_std = standarize_data(X, X_stats)
```

```python
X_std.head()
```

```python
# VIF test
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# First test
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X_std.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X_std.values, i)
                          for i in range(len(X_std.columns))]
# First test result
vif_data.astype(str)
vif_data_1 = vif_data.round(3)
```
