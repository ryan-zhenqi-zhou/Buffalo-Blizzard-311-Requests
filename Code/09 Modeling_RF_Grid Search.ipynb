```python
# Packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("../Data/modeling.csv")
print(len(df))
df.head()
```

```python
df_sample = df.sample(frac=0.7, axis=0, random_state=1)
df_sample.reset_index(drop=True, inplace=True)
```

```python
df_sample
```

```python
x_train = df_sample.set_index("Geography")
y_train = x_train.pop("311_index_per_property")
```

```python
x_train
```

```python
y_train = np.ravel(y_train)
```

```python
y_train
```

```python
len(x_train), len(y_train)
```

```python
number_demo = len(x_train.columns)
number_demo
```

```python
# Create list of max_features to be searched
max_features = [number_demo, int(number_demo/2), int(number_demo/3), 'sqrt', 'log2']
```

```python
# Grid search - sociodemographic features
rf = RandomForestRegressor(random_state=42)

pipe = Pipeline([
    ('scale', StandardScaler()), # standardization data set
    ('model', rf)
])

param_grid = [
    {'model__bootstrap':[True, False], 
     'model__n_estimators': range(10, 310, 10),
     'model__max_features':max_features,
     'model__max_depth': [5, 10, 15, 20]}
]

rf_search = GridSearchCV(pipe, param_grid, cv=10, scoring='neg_mean_squared_error', n_jobs=-1, return_train_score=True)
rf_search.fit(x_train, y_train)
```

```python
print(rf_search.best_params_)
```
